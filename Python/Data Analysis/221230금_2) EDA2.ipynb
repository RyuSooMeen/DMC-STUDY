> ### **데이터프레임 묶기**
> - concat과 merge 비교
---

```python
df1SE = pd.DataFrame({'StudentID':list(range(9,30,2)),
                           'ScoreSE' : [22,66,31,51,71,91,56,32,52,73,92]})
df2SE = pd.DataFrame({'StudentID':list(range(2,31,2)),
                           'ScoreSE' : [98,93,44,77,69,56,31,53,
                                      78,93,56,77,33,56,27]})
df1ML = pd.DataFrame({'StudentID':list(range(1,30,2)),
                           'ScoreML' : [39,49,55,77,52,86,41,77,
                                       73,51,86,82,92,23,49]})
df2ML = pd.DataFrame({'StudentID':list(range(2,21,2)),
                           'ScoreML' : [93,44,78,97,87,89,39,43,88,78]})
```

#### concat()

```python
dfSE = pd.concat([df1SE,df2SE],ignore_index=True)
dfML = pd.concat([df1ML,df2ML],ignore_index=True)
df = pd.concat([dfML,dfSE],axis = 1)
df
```

#### merge() inner

```python
dfSE = pd.concat([df1SE,df2SE],ignore_index=True)
dfML = pd.concat([df1ML,df2ML],ignore_index=True)

df= dfSE.merge(dfML,how='inner')
df
```

#### merge() left

```python
dfSE = pd.concat([df1SE,df2SE],ignore_index=True)
dfML = pd.concat([df1ML,df2ML],ignore_index=True)

df= dfSE.merge(dfML,how='left')
df
```

#### merge() right

```python
dfSE = pd.concat([df1SE,df2SE],ignore_index=True)
dfML = pd.concat([df1ML,df2ML],ignore_index=True)

df= dfSE.merge(dfML,how='right')
df
```

#### merge() outer

```python
dfSE = pd.concat([df1SE,df2SE],ignore_index=True)
dfML = pd.concat([df1ML,df2ML],ignore_index=True)

df= dfSE.merge(dfML,how='outer')
df
```

> ### **Reshaping**
> - stack/unstack
---

```python
data= np.arange(15).reshape((3,5))
indexers = ['Rainfall','Humidity','Wind']
dframe1 = pd.DataFrame(data,index=indexers,
                       columns=['Bergen','Oslo','Trondheim',
                                'Stavanger','Kristiansand'])
dframe1
```

```python
type(dframe1)
```

```python
data= np.arange(15).reshape((3,5))
indexers = ['Rainfall','Humidity','Wind']
dframe1 = pd.DataFrame(data,index=indexers,
                       columns=['Bergen','Oslo','Trondheim',
                                'Stavanger','Kristiansand'])
unstacked=dframe1.unstack()
print(unstacked)
print(type(unstacked))
print(unstacked.index)
```

```python
data= np.arange(15).reshape((3,5))
indexers = ['Rainfall','Humidity','Wind']
dframe1 = pd.DataFrame(data,index=indexers,
                       columns=['Bergen','Oslo','Trondheim',
                                'Stavanger','Kristiansand'])
stacked = dframe1.stack()
stacked
```

```python
stacked = dframe1.stack()
stacked
```

```python
unstacked=stacked.unstack()
unstacked
```

```python
series1 = pd.Series([0,111,222,333],
                    index = ['zeros','ones','twos','threes'])
series2 = pd.Series([444,555,666,],
                    index= ['fours','fives','sixs'])
frame2 = pd.concat([series1,series2],keys=['Number1','Number2'])
frame2
```

#### unstack의 default는 1

```python
unstacked=frame2.unstack(1)
unstacked
```

> ### **조건 만족하는 행 뽑기**
> - **nlargest, nsmallest**
> - **sort_values**
> - count, value_counts()
---

#### 가장 무거운 자동차 10개

```python
df.nlargest(10,'weight')
```

#### 가장 가벼운 자동차 10개 중, 엑셀이 가장 잘 나가는 차 5개

```python
df.nsmallest(10,'weight').nlargest(5,'acceleration')
```

#### 연식이 최근인 것들 중 mpg 순으로 정렬

```python
df.sort_values(['model year','mpg'],ascending=False)
```

#### 데이터의 열 마다 개수를 시리즈 형태로 변환
